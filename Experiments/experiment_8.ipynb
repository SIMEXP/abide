```python
import os
import glob
import numpy as np
import pandas as pd
import scipy as scp
import nibabel as nib
import brainbox as bb
from matplotlib import cm
import statsmodels.api as sm
from pymatbridge import Matlab
from matplotlib import pyplot as plt
import scipy.spatial.distance as dist
import scipy.cluster.hierarchy as clh
from sklearn.decomposition import FastICA, PCA
from nipy.modalities.fmri.glm import GeneralLinearModel
```

```python
# Paths
in_path = '/data1/abide/Out/Remote/all_worked/out/maps/stability_maps/correlation/scale_12'
data_path = os.path.join(in_path, 'netstack_net5.nii.gz')
pheno_path = os.path.join(in_path, 'matched_pheno.csv')
mask_in = os.path.join(in_path, 'mask.nii.gz')
```

```python
# Get the mask
m_img = nib.load(mask_in)
mask = m_img.get_data()
```

```python
# Get the network stack and mask it
n_img = nib.load(data_path)
n_data = n_img.get_data()
```

```python
# Get the subject by voxel representation of the data
n_subs = n_data.shape[3]
n_voxels = np.sum(mask > 0)
tseries = np.zeros((n_subs, n_voxels))
for ind in np.arange(n_subs):
    tmp = n_data[..., ind]
    tseries[ind, :] = tmp[mask > 0]
```

```python
# Regress the average signal across all subjets on the individual timeseries
mean_tseries = np.mean(tseries, axis=0)
n_subs = tseries.shape[0]
n_voxels = tseries.shape[1]
tseries_ga = np.zeros_like(tseries)
for s_id in np.arange(n_subs):
    mod = sm.OLS(tseries[s_id, :], mean_tseries)
    res = mod.fit()
    tseries_ga[s_id, :] = res.resid
```

```python
# Build a correlation matrix between subjects
R = np.corrcoef(tseries_ga)
linkage = clh.linkage(R, method='ward')
dend = clh.dendrogram(linkage, no_plot=True)
order = dend['leaves']
```

```python
def squareform(part):
    """
    visualize a partition as matrix
    """
    n = len(part)
    mat = np.zeros((n, n))
    for i_x, v_x in enumerate(part):
        for i_y, v_y in enumerate(part):
            if v_x == v_y:
                val = 1
            else:
                val = 0
            mat[i_x, i_y] = val
    return mat
```

```python
p_scale = 7
part = clh.fcluster(linkage, p_scale, criterion='maxclust')
```

```python
# Get the hierarchical partition
tmp = R[:, order]
R = tmp[order, :]
```

```python
p = squareform(part)
tmp = p[:, order]
p = tmp[order, :]
```

```python
plt.imshow(R, vmax=0.3, vmin=-0.3, cmap=cm.bwr)
plt.colorbar()
```

```python
# Get the cluster averages of the subjects
c_mat = np.zeros((p_scale, n_voxels))
for ind in np.arange(p_scale):
    c_mat[ind, :] = np.mean(tseries_ga[part==ind+1, :], axis=0)
```

```python
# map them back into the brain space
b_shape = n_data.shape[:3]
b_mat = np.zeros(b_shape + (p_scale,))
for ind in np.arange(p_scale):
    cam = c_mat[ind,:]
    b_mat[...,ind][mask > 0] = cam
```

```python
f = plt.figure(figsize=(20, 40))
for ind in np.arange(p_scale):
    ax = f.add_subplot(4, 2, ind+1)
    mat = bb.visuOps.make_montage(b_mat[..., ind])
    ax.imshow(mat, vmax=0.2, vmin=-0.2, cmap=cm.bwr)
```

```python
# Get the weights, Pierre style
weights = np.zeros((n_subs, p_scale))
for p_id in np.arange(p_scale):
    avg_cl = c_mat[p_id, ...]
    for s_id in np.arange(n_subs):
        sub = tseries_ga[s_id, ...]
        weights[s_id, p_id] = scp.stats.pearsonr(sub, avg_cl)[0]
```
