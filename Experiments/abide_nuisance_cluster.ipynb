```python
# Duplicate dropper
def drop_duplicates(in_dict):
    """
    Because python uses pointers and does not copy the variables
    I can operate directly on the dictionary and change it in place
    """
    cp_dict = copy.deepcopy(in_dict)
    subs = cp_dict['sub_name']
    dirs = cp_dict['dir']
    path = cp_dict['path']
    drop = list()
    present = list()
    sub_names = np.array([int64(re.search(r'(?<=\d{2})\d{5}', sub_id).group()) for sub_id in cp_dict['sub_name']])
    for idx, sub in enumerate(sub_names):
        if not sub in present:
            present.append(sub)
        else:
            drop.append(idx)
    print('Found {} items to drop'.format(len(drop)))
    # Pop them in reverse order
    for idx in drop[::-1]:
        subs.pop(idx)
        dirs.pop(idx)
        path.pop(idx)
    
    return cp_dict
```

## Define a couple of functions so we can iterate the process

```python
def perm(pheno_data, all_sub, n1, res, cov):
    pheno_all = pheno_data.loc[pheno_data['SUB_ID'].isin(all_sub)]
    level_list = list(pheno_all[cov].unique())
    perm_mat = np.zeros((len(level_list), res))
    perm_dict = dict()
    for ridx, it in enumerate(xrange(int(res))):
        ran = np.random.permutation(all_sub)
        ran1 = ran[:n1]
        ran2 = ran[n1:]
        # Get the pheno data for the randomized data again
        ranphen1 = pheno_data.loc[pheno_data['SUB_ID'].isin(ran1)]
        ranphen2 = pheno_data.loc[pheno_data['SUB_ID'].isin(ran2)]
        
        ran_count_1 = ranphen1[cov].value_counts()
        ran_count_2 = ranphen2[cov].value_counts()
        rlvl1 = list(ranphen1['SITE_ID'].unique())
        rlvl2 = list(ranphen2['SITE_ID'].unique())
        
        for lidx, level in enumerate(level_list):
            if not level in rlvl1 or not level in rlvl2:
                ratio = 0
            else:
                ran_num_1 = ran_count_1[level]
                ran_num_2 = ran_count_2[level]
                ratio = np.float(ran_num_1) / ran_num_2
                if ratio > 1:
                    ratio = 1/ratio
            perm_mat[lidx, ridx] = ratio
    for lidx, level in enumerate(level_list):
        perm_dict[level] = perm_mat[lidx, :]
    return perm_mat, perm_dict
```

```python
def emp(cp1, cp2, cov):
    sl1 = list(cp1[cov].unique())
    sl2 = list(cp2[cov].unique())
    tmp = list(set(sl1 + sl2))

    count_1 = cp1[cov].value_counts()
    count_2 = cp2[cov].value_counts()
    
    # Compute the empirical ratio
    emp_mat = np.zeros(len(tmp))
    emp_dict = dict()
    for lidx, level in enumerate(tmp):
        if not level in sl1 or not level in sl2:
            ratio = 0
        else:
            num_1 = count_1[level]
            num_2 = count_2[level]
            ratio = np.float(num_1) / num_2
            if ratio > 1:
                ratio = 1/ratio
        emp_mat[lidx] = ratio
        emp_dict[level] = ratio

    return emp_mat, emp_dict
```

```python
def analyze((clust1, clust2, pheno_data, data_subs, part, res)):
    # This should return the p-value and the matrix with the permuted
    # ratios
    # Do the empirical ratios first
    sub1 = data_subs[part==clust1]
    sub2 = data_subs[part==clust2]
    n1 = len(sub1)
    all_subs = np.concatenate((sub1, sub2), axis=0)
    cp1 = pheno_data.loc[pheno_data['SUB_ID'].isin(sub1)]
    cp2 = pheno_data.loc[pheno_data['SUB_ID'].isin(sub2)]
    emp_mat, emp_dict = emp(cp1, cp2, cov)
    # Now do the permutations
    perm_mat, perm_dict = perm(pheno_data, all_subs, n1, res, cov)
    
    return (emp_mat, perm_mat), (emp_dict, perm_dict)
```

```python
# Old FDR function
def computeFDR(pValueVector, alpha):
    # This returns a new thresholded p value
    # Sort the p values by size, beginning with the smallest
    sortedP = np.sort(pValueVector)
    # Reverse sort the p-values, so the first one is the biggest
    reverseP = sortedP[::-1]
    # Get the number of p-values
    numP = float(len(reverseP))
    # Create a vector designating the position of the reverse sorted p-values
    # in the sorted p vector (e.g. the first reverse sorted p-value will have
    # the index numP because it would be the last entry in the sorted vector)
    indexP = np.arange(numP, 0, -1)
    # Create test vector of (index of p value / number of p values) * alpha
    test = indexP / numP * alpha
    # Check where p-value <= test
    testIndex = np.where(reverseP <= test)
    if testIndex[0].size == 0:
        print('None of your p values pass FDR correction')
        pFDR = 0
    else:
        # Get the first p value that passes the criterion
        pFDR = reverseP[np.min(testIndex)]
        print('FDR corrected p value for alpha of {} is {}'.format(alpha, pFDR))

    return pFDR
```

# Visualize the findings

Now we have one cluster by cluster matrix for all the levels of SITE_ID. But for some cluster comparisons they did not exist and we don't have any values for these. How do we deal with them here? Maybe use -1 as a missing value.

```python
levels = pheno_data[cov].unique()
n_lvl = len(levels)
rows = 2
cols = np.floor(n_lvl / rows) + n_lvl%rows
f_row = 8 * rows
f_col = 10 * cols
all_mat = np.zeros((scale, scale))
# Now plot all comparisons of level 'CALTECH' for all cluster comparisons
f = plt.figure(figsize=(f_row,f_col))
for lidx, level in enumerate(levels):
    res_mat = np.zeros((scale, scale))
    # vector of unique p values for FDR
    p_l = list()
    for cidx, clust1 in enumerate(cl1):
        clust2 = cl2[cidx]
        result = parallel_results[cidx]
        r_e_d, r_p_d = result[1]
        if not level in r_e_d.keys():
            p = 2
        else:
            e_d = r_e_d[level]
            p_d = r_p_d[level]
            p = (np.sum(np.abs(p_d) > np.abs(e_d)) + 1) / float(res + 1)
            p_l.append(p)
        res_mat[clust1, clust2] = p
    # Flip the stuff
    res_mat[cl2, cl1] = res_mat[cl1, cl2]
    # Do FDR correction
    p_vec = np.array(p_l)
    p_corr = computeFDR(p_vec, 0.05)
    res_mat[res_mat > p_corr] = 0
    res_mat[res_mat != 0] = 1
    all_mat += res_mat
    
    # Visualize the results
    ax = f.add_subplot(cols, rows, lidx)
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.05)
    mp1 = ax.matshow(res_mat, cmap=plt.cm.Blues)
    ax.set_title('{}'.format(level))
    f.colorbar(mp1, cax=cax)
    
all_mat /= n_lvl
```
