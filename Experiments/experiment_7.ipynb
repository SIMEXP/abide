# Experiment 7

The purpose of this experiment is to compare the output of the python and niak approach

```python
import os
import glob
import numpy as np
import pandas as pd
import nibabel as nib
import brainbox as bb
from scipy import io as si
import pymatbridge as pymat
import statsmodels.api as sm
from matplotlib import pyplot as plt
from sklearn.decomposition import FastICA, PCA
```

```python
# Paths
in_path = '/data1/abide/Out/Remote/all_worked/out/maps/stability_maps/correlation/scale_12'
data_path = os.path.join(in_path, 'netstack_net10.nii.gz')
pheno_path = os.path.join(in_path, 'matched_pheno.csv')
mask_in = os.path.join(in_path, 'mask.nii.gz')
```

```python
m_img = nib.load(mask_in)
mask = m_img.get_data()
```

```python
# Get the network stack and mask it
n_img = nib.load(data_path)
n_data = n_img.get_data()
# Get the subject by voxel representation of the data
n_subs = n_data.shape[3]
n_voxels = np.sum(mask > 0)
mat = np.zeros((n_subs, n_voxels))
for ind in np.arange(n_subs):
    tmp = n_data[..., ind]
    mat[ind, :] = tmp[mask > 0]
```

```python
mld = si.loadmat(os.path.join(in_path, 'matlab_ica_components.m'))
```

```python
com_m = mld.values()[0]
```

```python
# Run an ICA on the data
n_cmp = 10
ica = FastICA(n_components=10)
results = ica.fit_transform(mat)
```

```python
comps = ica.components_
# And map them back into the brain space
b_shape = n_data.shape[:3]
c_mat = np.zeros(b_shape + (n_cmp,))
for ind in np.arange(n_cmp):
    cm = comps[ind,:]
    c_mat[...,ind][mask > 0] = cm
```

```python
f = plt.figure(figsize=(20, 100))
for ind, pos in enumerate(np.arange(1,20,2)):
    ax_p = f.add_subplot(10, 2, pos)
    mat_p = bb.visuOps.make_montage(c_mat[..., ind])
    ax_p.imshow(mat_p, vmin=0)
    ax_p.set_title('python component {}'.format(ind+1))
    ax_m = f.add_subplot(10, 2, pos + 1)
    mat_m = bb.visuOps.make_montage(com_m[..., ind])
    ax_m.imshow(mat_m, vmin=0)
    ax_m.set_title('matlab component {}'.format(ind+1))
```
